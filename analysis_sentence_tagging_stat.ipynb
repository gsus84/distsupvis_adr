```python
import os
import json

from typing import Dict
from typing import Any

from collections import Counter
from dataclasses import dataclass
from dataclasses import field

from src.medline_preprocessing.medl_sent_parse import MedlineTagStatistic
```

```python
@dataclass
class TagSentsStats:
    total_tag_stat: MedlineTagStatistic = MedlineTagStatistic()
    # single_file_tag_stats: Dict[str, Any] = field(default_factory=dict)

    def get_all_sent_tag_stats(self, path: str = "data/statistic/tag_stats/"):
        for stat_file in os.listdir(path):
            with open(path + stat_file, "r") as json_file:
                stat_file_data = json.load(json_file)
            stat_file_data = MedlineTagStatistic(**stat_file_data)
            self.total_tag_stat += stat_file_data
            # self.single_file_tag_stats[stat_file] = stat_file_data
    
```

```python
tag_stat = TagSentsStats()
tag_stat.get_all_sent_tag_stats()
```
